```python
import pandas as pd
import numpy as np
import glob
from IPython.display import display, HTML
```

```python
files_list = glob.glob('other_minerals_*_*.csv')
files_list
```

```python
col_names = ['Code/Ref','Company_Name_cl','License_type_cl','Application_date_cl','Award_date_cl', \
             'Expiry_date_cl','Commodity_cl','Area_km2_cl','Area_acre_cl','Location_state_cl', \
             'License_location_cl','License_location_code_1_cl','License_location_code_2_cl']

dataset_2014_15 = pd.DataFrame(columns=col_names)
dataset_2015_16 = pd.DataFrame(columns=col_names)
```

```python
def me2_match_locations(df):
    lookup_sheet = pd.read_csv('ME2_2014-15_state_locations.csv')
    lookup_sheet.columns = lookup_sheet.columns.str.replace('ï»¿Code/Ref','Code/Ref')
    lookup_sheet = lookup_sheet[['Code/Ref','Location (Region)']]
    df = df.merge(lookup_sheet,on='Code/Ref')
    
    state_col = 'Location (Region)'
    
    df[state_col] = df[state_col].str.replace('State','')
    df[state_col] = df[state_col].str.strip()
    df[state_col] = df[state_col].str.title()
    df.loc[df[state_col]  == 'Taninthayi', state_col] = 'Tanintharyi'
    df.loc[df[state_col]  == 'Kayar', state_col] = 'Kayah'
    df.loc[df[state_col]  == 'Shan (S)', state_col] = 'Shan'
    df.loc[df[state_col]  == 'Mandalay+Shan', state_col] = ''
    
    if "Location_state_cl" not in df.columns:
        df["Location_state_cl"] = np.nan
    
    df["Location_state_cl"] = df["Location_state_cl"].fillna(df[state_col])
    
    return df
```

```python
for f in files_list:
    df = pd.read_csv(f)
    print("---"+f+"---")
    year = f.split('_')[2]
    department = f.split('_')[3].split('.csv')[0]
    print(df.columns)
    for c in df.columns:
        if ("_cl" not in c and "Code" not in c):
            df.drop(c, axis=1,inplace=True)
    if 'Code / Ref' in df.columns:
        df.rename(columns={"Code / Ref": "Code/Ref"}, inplace=True)
    if 'Company_name_cl' in df.columns:
        df.rename(columns={"Company_name_cl": "Company_Name_cl"}, inplace=True)
    df['Year'] = year
    df['Department'] = department
    
    if department == 'ME2':
        df = me2_match_locations(df)
    
    display(df.head())
    
    #display(df['License_type_cl'].str.replace(' ','|').value_counts())
    
    if (year == '2014-15'):
        dataset_2014_15 = dataset_2014_15.append(df, ignore_index=True)
    else:
        dataset_2015_16 = dataset_2015_16.append(df, ignore_index=True)
```
